```python
import random

user = input("Please enter the correct option (rock, paper, scissor): ").lower()
options = ["rock", "paper", "scissor"]
user_win = 0
computer_win = 0

while user_win < 5 and computer_win < 5:
    computer = random.choice(options)
    print(f"Computer chose: {computer}")

    if user == computer:
        print("It's a tie!")
    elif (user == "rock" and computer == "scissor") or \
         (user == "scissor" and computer == "paper") or \
         (user == "paper" and computer == "rock"):
        print("You win this round!")
        user_win += 1
    else:
        print("Computer wins this round!")
        computer_win += 1

    print(f"Score: You {user_win} - {computer_win} Computer\n")


    user = input("Please enter the correct option (rock, paper, scissor): ").lower()

if user_win > computer_win:
    print("Congratulations! You are the final winner!")
else:
    print("Sorry, the computer is the final winner.")
```
